```python
import pandas  as pd
from collections import defaultdict
from matplotlib.pyplot import figure
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
dataset_names = ['abalone', 'epileptic_seizure', 'HARUS', 'online_shoppers_intention', 'polish_companies_bankruptcy']
dataset_names_style = ['Abalone', 'ESR', 'HARUS', 'OSPI', 'PCB']
n_estimators = [10,20,40,60,80,100]

colors_mcs = ["#e41a1c","#377eb8","#4daf4a","#984ea3","#ff7f00","#000000"]
markers = ["o-","^-","s-","D-","*-","X-","P-"]
models_names = {'bagging_oracle': 'Bagging', 'adaboost_oracle': 'Adaboost', 'randomsubspace_oracle': 'Random Subspace', 'randomlinearoracle_oracle': 'RLO' }
```

```python
count = 0
for dataset in dataset_names:    
    models_acc_mean = defaultdict(list)
    low_value = 100
    for qtd_es in n_estimators:
        df = pd.read_csv(dataset +'/'+dataset+'_model_score_'+str(qtd_es)+ '_estimators', sep=',')
        for index, row in df.iterrows():
            models_acc_mean[row['model']].append(row['mean'])
            if((row['mean']*100) < low_value):                
                low_value = row['mean'] *100
    
    with sns.color_palette("tab10"):  
        figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
        plt.xlabel('Tamanho do pool - ' + dataset_names_style[count], fontsize=15)       
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.ylabel('Oracle Score %', fontsize=15)
        i = 0
        for key, value in models_acc_mean.items():
            lista = [x * 100 for x in value]
            plt.plot(n_estimators, lista,markers[i], label=models_names[key],c=colors_mcs[i], markersize=10)
            i += 1
        #plt.yticks(np.arange(low_value-1, 101, (101 - low_value-1)/10)) 
        plt.xticks([0,10,20,40,60,80,100]) 
        plt.legend()
        plt.show()
        count += 1
```
