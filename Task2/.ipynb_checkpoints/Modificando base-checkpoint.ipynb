```python
import pandas as pd


cm1_data = pd.read_csv("cm1.csv", header=None)
```

```python
cm1_data
```

```python
cm1_n_row, cm1_n_column = cm1_data.shape
cm1_columns = ["X{}".format(i) for i in range(0, cm1_n_column-1)] + ["y"]
cm1_data.columns = cm1_columns
cm1_data["y"] = cm1_data["y"].apply(lambda x: 1 if x == True else 2)
cm1_columns = cm1_columns[-1:] + cm1_columns[:-1]
cm1_data = cm1_data[cm1_columns]
```

```python
cm1_data
```

```python
cm1_data.to_csv('new_cm1.csv', encoding='utf-8', index=False, header=False)
```
