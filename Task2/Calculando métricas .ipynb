```python
import pandas as pd
from collections import defaultdict
import numpy as np
```

```python
headers = ['FOLD', 'q_init', 'q_v', 'q_hard', 'q_easy', 'acc_init', 'acc_v', 'acc_hard', 'acc_easy', 
                'auc_init', 'auc_v', 'auc_hard', 'auc_easy', 'g_mean_init', 'g_mean_v', 'g_mean_hard', 'g_mean_easy',
                'f1_init', 'f1_v', 'f1_hard', 'f1_easy', 'size_init', 'size_v', 'size_hard', 'size_easy' ]

metrics = ['q', 'acc', 'auc', 'g_mean', 'f1', 'size']
validations = ['init', 'v', 'easy', 'hard']
```

```python


cm1_data = pd.read_csv("cm1_result_fold.txt", names=headers)
cm1_data = cm1_data.fillna(0)
cm1_data
```

```python
kc1_data = pd.read_csv("kc1_result_fold.txt", names=headers)
kc1_data = kc1_data.fillna(0)
kc1_data
```

```python
def generate_mean_std(data):
    retorno = defaultdict(tuple)
    for index, row in data.items():
        retorno[index] = (np.mean(row), np.std(row))
        
    return retorno

metrics_cm1 = generate_mean_std(cm1_data)
metrics_kc1 = generate_mean_std(kc1_data)
```
