```python
import pandas  as pd
import numpy as np
import matplotlib
import seaborn as sns
```

```python
dataset_dir = 'results_kc1/'

metrics = ['accs', 'aucs','f1s','gm1']
metrics_names = {'accs': 'Acur√°cia', 'aucs': 'AUC', 'f1s': 'F1-Score', 'gm1': 'G-Mean'}
```

```python
f_size = 14

def plot_metric(metric,df):
    sns.set_style("white")
    ax = df.plot(kind="bar",title=metrics_names[metric], figsize=(10,4),fontsize=12)
    ax.set_xticklabels(['V', 'V-Hard', 'V-Easy'] ,rotation=0)
    ax.set_ylabel('Valor (%)' , fontsize=f_size)
    ax.set_xlabel('Conjunto de Valida√ß√£o' , fontsize=f_size)
    ax.set_axisbelow(True)
    ax.grid(True)

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(True)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(True)    
    ax.legend(bbox_to_anchor=(1.0, 1), loc='upper left', borderaxespad=0.)




for metric in metrics:

    df = pd.read_csv(dataset_dir + metric, sep=',')

    df.columns[0]
    df.set_index(df.columns[0], inplace=True)
    df = df.multiply(100)
    plot_metric(metric, df)
```
